```python
import csv
f=open('./data/seoul.csv') #seoul.csv 파일을 읽기 모드로 블러옴
data=csv.reader(f)
header=next(data)          #처번째 행(제목행) header 변수에 저장


max_temp=0
max_day=""

for row in data:
    if row[4]>"":
        if float(row[4])>max_temp:
            max_temp=float(row[4])
            max_day=row[2]
            
            
print(f'날짜: {max_day}, 최고기온 {max_temp}')

f.close()
```

```python
# plt 차트에서 한글 지원 방법 1
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path='C:/Windows/Fonts/H2HDRM.TTF'             #한글 폰트 경로 및 이름 정의
font_name=fm.FontProperties(fname=font_path).get_name() #폰트 속성 변경
plt.rc('font', family=font_name)              #차트 전체에 포트 속성 변경

plt.plot([33,13,22,9,45], label='데이터', color='skyblue' ,marker="o")      #1번 차트 및 옵션
plt.plot([50,40,30,20,10],'pink', label='cba',marker="^")     #2번 차트 및 옵션

plt.title("차트 그리기 실습")           #차트 제목
plt.xlabel("X-Label")          #x축 제목
plt.ylabel("Y-Label")          #y축 제목
plt.legend()                   #범례 출력


plt.show()
```

```python
# plt 차트에서 한글 지원 방법 2
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path='C:/Windows/Fonts/H2HDRM.TTF'             #한글 폰트 경로 및 이름 정의
font_name_1=fm.FontProperties(fname=font_path, size=18) #폰트 속성 변경
font_name_2=fm.FontProperties(fname=font_path, size=13)

plt.plot([33,13,22,9,45], label='데이터', color='skyblue' ,marker="o")      #1번 차트 및 옵션
plt.plot([50,40,30,20,10],'pink', label='cba',marker="^")     #2번 차트 및 옵션

plt.title("차트 그리기 실습" ,fontproperties=font_name_1)           #차트 제목
plt.xlabel("X-Label",fontproperties=font_name_2)          #x축 제목
plt.ylabel("Y-Label",fontproperties=font_name_2)          #y축 제목
plt.legend()                   #범례 출력


plt.show()
```

```python
#['\t\t108', '서울', '1907-10-01', '13.5', '20.7', '', '7.9', '12.8']
#날짜데이터(row[2])를 가져와 '-'을 기준으로 글자 나누기
#생일에 따른 년도별 최고기온/최저기온 차트 출력
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path='C:/Windows/Fonts/H2HDRM.TTF'   

import csv
f=open('./data/seoul.csv') #seoul.csv 파일을 읽기 모드로 블러옴
data=csv.reader(f)
header=next(data)          #처번째 행(제목행) header 변수에 저장
 
max_temp=[]   #최고기온
min_temp=[]   #최저기온
year_temp=[]  #생일자 년도

for row in data:
    if row[2]!="":
        if int(row[2].split('-')[0])>=1993:     # 1993년 이후 데이터 검색
            if row[2][-5:]=="02-19":            #2월 19일 데이터만 추출
                max_temp.append(float(row[4]))
                min_temp.append(float(row[-2]))
                year_temp.append(row[2].split('-')[0])

#데이터를 이용한 시각화
 
plt.figure(figsize=(14,5)) #차트 크기설정
plt.plot(year_temp, max_temp,label="최고기온", color='r',marker='o')
plt.plot(year_temp, min_temp,label="최저기온", color='b',marker='o')

plt.xticks(size=10, rotation=45) # x축 라벨에 대한 회전, 글자크기 vertical, 15, 30 ,90 


plt.title("내 생일 기온 변화, size=20")
plt.xlabel('년도' ,size='15')
plt.ylabel('온도', size='15')
plt.legend(loc="best")  #best, lower right ...
plt.grid(True)          # 그림 영역에 그리드 출력



plt.show()
```

### 미션: 평균데이터를 이용한 날짜별 차트 출력
- 1960년 이후 8월 15일 데이터를 기준으로 평균기온을 이용해 차트 작성
- 1960년 이후 1월 15일 데이터를 기준으로 평균기온을 이용해 차트 작성
- 기간: 1960~2020년 까지

```python
import csv
f=open('./data/seoul.csv') 
data=csv.reader(f)
header=next(data)          
 
mid_temp1=[]
mid_temp2=[]
year_temp=[]

for row in data:
    if row[2]!="":
        if int(row[2].split('-')[0])>=1960 and int(row[2].split('-')[0])<=2020:      
            if row[2][-5:]=="08-15":            
                mid_temp1.append(float(row[3]))
                
            if row[2][-5:]=="01-15":            
                mid_temp2.append(float(row[3]))
                year_temp.append(row[2].split('-')[0]) 
              
#데이터를 이용한 시각화
 
plt.figure(figsize=(14,5)) #차트 크기설정
plt.plot(year_temp, mid_temp1,label="8월 15일 평균기온", color='r',marker='o')
plt.plot(year_temp, mid_temp2,label="1월 15일 평균기온", color='b',marker='o')

plt.xticks(size=5, rotation=45) # x축 라벨에 대한 회전, 글자크기 vertical, 15, 30 ,90 


plt.title("1960년도 8월15일 평균기온", size=20)

plt.xlabel('년도' ,size='15')
plt.ylabel('온도', size='15')
plt.legend(loc="best")  
plt.grid(True)          



plt.show()
```
